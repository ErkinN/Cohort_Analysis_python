```python
import pandas as pd
import numpy as np
from datetime import datetime
```

```python
file_path=("C:/Users/User/Downloads/KPMG_VI_New_raw_data_update_final.xlsx/KPMG_VI_New_raw_data_update_final.xlsx")
df=pd.read_excel(file_path,sheet_name='Transactions')
#This line sets the column names of df1 using the first row (i.e., the 0th index) of the DataFrame
df.columns=df.iloc[0]
df.drop(df.index[0],inplace=True,axis=0)

#This line filters the DataFrame to only include rows where the order_status column has the value 'Approved'
df=df[df['order_status']=='Approved']
df
```

```python
#This line removes the first row of the DataFrame because it was used as the column headers in the previous step.
df_new=df[['customer_id','transaction_date']].copy(deep=True)
df_new
```

```python
#The code converts the transaction_date in each row into a year-month format and stores it as an integer in the new 'YM' column
df_new['YM']=df_new.apply(lambda x: int(datetime.strftime(x.transaction_date, "%Y%m")), axis=1)
df_new
```

```python
# This code finds the earliest year and month in the 'YM' column of df2 and stores that value in the variable start_month.
start_month=min(df_new.YM)
start_month
```

```python
#This code calculates the difference between each YM value and the start_month, effectively creating a month 
# index where the transactions in the start month will have an index of 0, 
# and each following month will have an index that increases by the difference in months
df_new['transaction_month_index']=df_new.YM.apply(lambda x: x-start_month)
df_new
```

```python
#This code groups the data by each customer_id, finds the earliest transaction month for each customer (using the minimum of the transaction_month_index),
#and then resets the index to get a DataFrame.
df1 = df_new.groupby('customer_id')['transaction_month_index'].min().reset_index()
df1 = df1.rename(columns={'transaction_month_index':'cohort_month'})
df1
```

```python
df2 = df_new.join(df1.set_index('customer_id'), on='customer_id')
df2
```

```python
df2['cohort_index'] = df2.transaction_month_index - df2.cohort_month
df2
```

```python
df3 = df2[['customer_id','cohort_month','cohort_index']]
df3
```

```python
df4 = df3.drop_duplicates(subset=['customer_id', 'cohort_month','cohort_index'], keep='first')
df4
```

```python
#This code groups the data by each combination of cohort_month and cohort_index, counts the number of unique customers in each group, 
# and then resets the index. 
#Finally, it renames the count column to customer_count.
df5 = df4.groupby(['cohort_month','cohort_index'])['customer_id'].count().reset_index()
df5 = df5.rename(columns={'customer_id':'customer_count'})
df5
```

```python
# This code creates a pivot table from df5 that summarizes the count of customers 
# for each cohort month and cohort index, effectively restructuring the data for easier analysis.
df6 = df5.pivot_table(index = 'cohort_month', columns = 'cohort_index',values = 'customer_count')
df6
```

```python
# This code calculates the percentage of customers in each cohort month and index relative to the number of customers in the 
# first cohort index (cohort_index = 0), effectively providing a retention rate analysis.
df7 = df6.divide(df6.iloc[:,0],axis=0).round(3)*100
df7
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

fig, ax = plt.subplots(1,1,figsize = (16,10))

sns.heatmap(df6, annot = True, cmap="YlGnBu", fmt='g')

ax.set_xlabel('Cohort Index')
ax.set_ylabel("Cohort Month")
ax.xaxis.set_label_position('top')

plt.show() 
```
